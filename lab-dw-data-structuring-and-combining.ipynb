This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
data_set_marketing = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
data_set_marketing.drop(columns=['unnamed:_0'], inplace=True)
display(data_set_marketing)
```

```python
data_set_marketing.isna().sum()
```

```python
data_set_marketing['sales_channel'].value_counts()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
sales_channel_total = data_set_marketing.groupby('sales_channel').agg({'income':sum}).round(2)
display(sales_channel_total)
#as there are no "mail" results, it won't appear on our aggregation table.
```

```python
sales_channel_total.sum()
```

```python
agent_percent = (152490152*100)/409520867
branch_percent = (113775608*100)/409520867
call_center_percent = (81055004*100)/409520867
web_percent = (62200103*100)/409520867
print("This is the percentage from Agent, compared to total income", round(agent_percent, 2))
print("This is the percentage from Branch, compared to total income", round(branch_percent,2))
print("This is the percentage from Call Center, compared to total income", round(call_center_percent,2))
print("This is the percentage from Web, compared to total income", round(web_percent,2))
```

```python
#were we can see a strong dominance on Agent and Branch, being +50% of the income. 
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
avg_cust_lifetime = data_set_marketing.groupby(['education', 'gender']).agg({'customer_lifetime_value' : 'mean'}).round(2)
display(avg_cust_lifetime)
```

```python
#in this table we can see that the mean per gender and education it's really similar, and there are no high differences. 
#Most difference we can see in Collegue, where male has a slight predominance comparing to female. 
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
month_complaints_policy_type = data_set_marketing.groupby(['month', 'policy_type']).agg({'number_of_open_complaints' : sum}).reset_index().round(2)
display(month_complaints_policy_type)
```

```python
long_format = pd.melt(month_complaints_policy_type, id_vars=['month','policy_type'])
long_format.head()
long_format.rename(columns={'value':'number_of_open_complaints'})
```
